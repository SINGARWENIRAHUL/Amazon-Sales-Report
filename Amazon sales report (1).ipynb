```python
# Data Cleaning and Preprocessing
# Convert 'date' column to datetime format for time series analysis
sales['Date'] = pd.to_datetime(sales['Date'], errors='coerce')
sales['Date']
```

```python
# Handle missing values (you can choose to remove or impute missing values)
sales.dropna(subset=['Amount'], inplace=True)  # Drop rows with missing important columns
sales['Amount'] = sales['Amount'].fillna(sales['Amount'].median())  # Impute missing values in 'amount' with the median
sales['Amount']
```

# 1. Sales Overview Analysis

```python
# Total sales
total_sales = sales['Amount'].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
# Sales over time (monthly trend)
sales['month'] = sales['date'].dt.to_period('M')
monthly_sales = sales.groupby('month')['Amount'].sum()
monthly_sales
```

```python
# Plot sales over time
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', color='b', marker='o')
plt.title('Sales Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

# 2. Product Analysis

```python
# Sales by  Category
category_sales = sales.groupby('Category')['Amount'].sum().sort_values(ascending=False)
category_sales
```

```python
# Plot product category sales
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar', color='g')
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top-selling sizes (if applicable)
size_sales = sales.groupby('Size')['Amount'].sum().sort_values(ascending=False)
size_sales
```

```python
plt.figure(figsize=(10, 6))
size_sales.plot(kind='bar', color='c')
plt.title('Sales by Size')
plt.xlabel('Size')
plt.ylabel('Sales Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

# 3. Fulfillment Analysis

```python
Fulfilment_sales = sales.groupby('Fulfilment')['Amount'].sum().sort_values(ascending=False)
Fulfilment_sales


```

```python
# Plot fulfillment method sales
plt.figure(figsize=(8, 6))
Fulfilment_sales.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Sales Distribution by Fulfilment Method')
plt.ylabel('')  # Remove the y-label as it's not needed for pie chart
plt.show()
```

# 4. Customer Segmentation

```python
# Assuming 'customer_id' exists, segment by customer frequency
Customer_sales = sales.groupby('Order ID')['Amount'].sum()
Customer_frequency = sales.groupby('Order ID').size()
Customer_sales
```

```python
Customer_frequency
```

```python
# Plot customer frequency distribution 
plt.figure(figsize=(10, 6))
sns.histplot(Customer_frequency, kde=True, color='orange')
plt.title('Customer Purchase Frequency Distribution')
plt.xlabel('Number of Purchases per Customer')
plt.ylabel('Number of Customers')
plt.show()
```

# 5. Geographical Analysis

```python
# Assuming the dataset has 'state' and 'city' columns, analyze sales by state
State_sales = sales.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)
State_sales
```

```python
city_sales = sales.groupby('ship-city')['Amount'].sum().sort_values(ascending=False)
city_sales
```

```python
# Plot geographical sales (by state)
plt.figure(figsize=(30, 6))
State_sales.plot(kind='bar', color='purple')
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Sales Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

# 6. Business Insights and Recommendations

```python
# Top-selling products (Top N products by revenue)
top_n_products = sales.groupby('Category')['Amount'].sum().nlargest(10)
print("Top 10 Selling Category by Revenue:")
print(top_n_products)
```

```python
# Sales recommendations
# High sales channels to focus on
sales_channel_sales = sales.groupby('Sales Channel')['Amount'].sum().sort_values(ascending=False)
print("\nSales by Sales Channel:")
print(sales_channel_sales)
```

```python
# Potential action for improving fulfillment efficiency
Fulfilment = sales.groupby('Fulfilment')['Status'].value_counts()
print("\nFulfilment Method Efficiency (Order Status Breakdown):")
print(Fulfilment)
```
